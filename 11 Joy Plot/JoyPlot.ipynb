```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import joypy
%matplotlib inline
```

```python
# 创建示例数据
np.random.seed(10)
data = pd.DataFrame(np.random.normal(loc=0, scale=1, size=(100, 5)), columns=list('ABCDE'))

# 创建 Joyplot 图
fig, ax = joypy.joyplot(data, figsize=(8, 6))

# 添加标题和标签
plt.title("Joyplot Example")
plt.xlabel("Value")
plt.ylabel("Density")

# 显示图形
plt.show()
```

```python

# 创建示例数据
np.random.seed(10)
data = pd.DataFrame(np.random.normal(loc=0, scale=1, size=(100, 5)), columns=list('ABCDE'))
# 在数据框中添加一个分组列
data['Group'] = np.random.choice(['Group1', 'Group2'], size=len(data))
# 创建 Joyplot 图
fig, ax = joypy.joyplot(data, 
                        column=['A', 'B', 'C','D','E'], 
                        by = 'Group',
                        grid=True,
                        xlabelsize=10,
                        ylabelsize='medium',
                        xrot=30,
                        yrot=None,
                        hist=False,
                        fade=True,
                        ylim='max',
                        fill=True,
                        linecolor='blue',
                        background=None,
                        overlap=1,
                        title=None
                        #colormap=plt.cm.Blues_r
                        )

# 添加标题和标签
plt.title("Joyplot Example")
plt.xlabel("Value")
plt.ylabel("Density")

# 显示图形
plt.show()
```
